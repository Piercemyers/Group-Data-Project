```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv("../input/mushroom-classification/mushrooms.csv")
df.head()
```

```python
#Get to columns to be compared

X = df[['class','cap-shape']]
X=pd.get_dummies(X,prefix_sep = ',')
X.head(2)

```

```python
#Cap Shape
#Wittle down correlations further to properly compare the two relevant columns

plt.subplots(figsize=(20,15))
corr = X.corr()
sub_corr = corr[['cap-shape,b','cap-shape,c', 'cap-shape,f','cap-shape,k','cap-shape,s','cap-shape,x']]


sub_corr.head(2)
ax = sns.heatmap(sub_corr.iloc[0:2],#Gets relevant rows
                 vmin=-1, vmax=1, center=0,#Set the key properly
                 cmap=sns.diverging_palette(20, 220, n=200),square=True, annot = True) #Set colors

ax.set_title('Correlation between Mushroom Cap Shape and Edibility')
ax.set_ylabel('Edibility')
ax.set_xlabel('Cap Shape')

```
